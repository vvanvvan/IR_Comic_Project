```python
csv_reader = csv.reader(file)
```

สร้างlistว่าง เพื่อจะทำการเก็บข้อมูลในแต่ละcolumnไว้ในlist

```python
name_of_comic=[]
synopsis=[]
genre=[]
original_type=[]
rating=[]
```

วนลูปเพื่อเอาข้อมูลจากcsv ไปใส่ในlist


```python
for column in csv_reader:
  name_of_comic.append(column[0])
  synopsis.append(column[1])
  genre.append(column[2])
  original_type.append(column[3])
  rating.append(column[4])
```

กำหนดช่วงของlistใหม่ เพื่อลบcolumn headออก

```python
name_of_comic = name_of_comic[1:]
synopsis = synopsis[1:]
genre = genre[1:]
original_type = original_type[1:]
rating = rating[1:]
```

# TOKENIZATION
import library 
-  import library nltk และdownload ฐานข้อมูลของstopword
-  import library RegexpTokenizer เพื่อใช้ทำ word token


```python
import nltk
nltk.download('stopwords')

from nltk.tokenize import RegexpTokenizer
```

สร้างตัวแปรlistเพื่อเอาไว้เก็บข้อมูล เมื่อทำการtokenสำเร็จแล้ว (ไม่ทำtoken rating)

```python
stw_name_of_comic=[]
stw_synopsis=[]
stw_genre=[]
stw_original_type=[]
```

ใช้ RegexpTokenizer("[\w']+") เพื่อทำการtoken และทำให้เป็นlowercase

```python
tokenizers = RegexpTokenizer("[\w']+")
```

ในขั้นตอนถัดไปจะเป็นการทำWord Tokenในแต่ละColumn

โดยจะทำ Word Token -> ชื่อcomic, ประเภทของcomic และภาษาต้นฉบับของComic

ส่วนsysnopsis หรือ เรื่องย่อ จะทำToken, Stopword, Lemmatizerพร้อมกันเลย

```python
for comic in name_of_comic:
     stw_name_of_comic.append(tokenizers.tokenize(comic))
     #print(tokenizers.tokenize(comic))
```

```python
for comic in genre:
    stw_genre.append(tokenizers.tokenize(comic))
    #print(tokenizers.tokenize(comic))
```

```python
for comic in original_type:
    stw_original_type.append(tokenizers.tokenize(comic))
    #print(tokenizers.tokenize(comic))
```

```python
import nltk
nltk.download('wordnet')

from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
 
lem = WordNetLemmatizer()

stopword=set(stopwords.words('english'))
tokenizers = RegexpTokenizer("[\w']+")

serie_story =[]
for comic in synopsis:

     #Token word
     word_tokens = tokenizers.tokenize(comic)
     
     #Stopword
     filtered_list = [w for w in word_tokens if not w.lower() in stopword]
     
     
     #Lemmatization (สนใจหลักไวยกรณ์)
     lemmatized_output = ' '.join([lem.lemmatize(w) for w in filtered_list])
     
     stw_synopsis.append(tokenizers.tokenize(lemmatized_output))
     #print('Lemmatization: '+str(tokenizers.tokenize(lemmatized_output)))   
```

สร้างตัวแปรlist เพื่อเอาไว้เก็บค่า และวนลูปให้ทุกColumnมาอยู่ในlistเดียวกัน เพื่อจะได้นำไปทำขั้นตอนถัดไปต่อได้

```python
keys = []
for w in range(200):
  comic = stw_name_of_comic[w] + stw_name_of_comic[w] + stw_synopsis[w] + stw_original_type[w] + stw_genre[w]
  print(comic)
  keys.append(comic)

#print(keys)
```

เก็บข้อมูลลงในcsv

```python
imdb_df = pd.DataFrame ({
    'comic' : keys,
    })


imdb_df.to_csv('comic_end.csv')
```
