# QUERY


ในขั้นตอนถัดไปจะเป็นการทำ(Query)

*   word token
*   stopword
*   ทำเป็นรากศัพท์ โดยใช้ Lemmatization (สนใจไวยกรณ์)
*   คำนวณค่า WEIGHT = TF * IDF
*   คำนวณค่า Cosine Similarity
*   แสดงTop Rankจากค่า Cosine Similarity


import library nltk และdownload 

*   stopwords ใชในการทำstopword
*   wordnet ใชในการทำlemma



```python
import nltk
#import pickle

#nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
```

สร้างฟังก์ชันเพื่อเอาไว้ทำQuery -> Word Token

```python
def get_token(doc_text):
  tokenizer = nltk.RegexpTokenizer("[\w']+")
  tokens = tokenizer.tokenize(doc_text)
  return tokens
```

สร้างฟังก์ชันเพื่อเอาไว้ทำQuery -> Stop Word

```python
stopword=set(stopwords.words('english'))
def remove_stopword(doc_text):
  cleaned_text=[]
  for words in doc_text:
    if words not in stopword:
      cleaned_text.append(words)
  return cleaned_text   
```

สร้างฟังก์ชันเพื่อเอาไว้ทำQuery -> Lemma

```python
def get_lemma(token_list):
  lem = nltk.stem.WordNetLemmatizer()
  lemma=[]
  for words in token_list:
    lemma.append(lem.lemmatize(words))
  return lemma
```

# Type Query ❗
ใส่query ที่ต้องการเพื่อทำการสืบค้น

```python
query = 'romantic king and queen'
```

นำqueryส่งไปฟังก์ชันด้านบน เพื่อทำword token, stopword และlemma

```python
query=get_token(query)
query=remove_stopword(query)
#print(query)
q=[]
for w in get_lemma(query):
  q.append(w)
q=' '.join(q)
```

Query ที่ผ่านการทำword token, stopword และlemma ✅

```python
print(q)
```

# คำนวณ WEIGHT = TF * IDF
ใช้ TfidfVectorizer จาก sklearn library เพื่อคำนวณ WEIGHT = TF * IDF

```python
query_1 = vectorizer.transform([q])
df2 = pd.DataFrame(query_1.toarray(), columns=vectorizer.get_feature_names_out())

df2
```

# คำนวณค่าCosine Similarity 🔎
คำนวณค่าCosine SimilarityจากDoc และ Query โดยจะprintลำดับtop rank ออกมา 10 อันดับ

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity = cosine_similarity(doc_1,query_1).flatten()
rank = cosine_similarity.argsort()[:-11:-1]
print(rank)
```

อ่านไฟล์จากCSVที่ยังไม่ได้ทำword token, stopword และlemma เพื่อนำมาแสดงข้อมูล

```python
name_of_comic=[]
synopsis=[]
genre=[]
original_type=[]
rating=[]
for row in csv_reader:
  name_of_comic.append(row[1])
  synopsis.append(row[2])
  genre.append(row[3])
  original_type.append(row[4])
  rating.append(row[5])
```

```python
top_rank = []
top_rank2 = []

top_comic1 = []
top_comic2 = []
top_comic3 = []
top_comic4 = []
top_comic5 = []

show_rank = 1
for i in rank:

    data = [the_comics[i]]
    print("อันดับที่", show_rank)
    top_rank2.append(show_rank) 

    print("ค่า cosine similarity = ",cosine_similarity[i])
    show_rank = show_rank + 1
    #print(data)
    print(data)
    print(" ")

    top_rank.append(cosine_similarity[i])
 
    top_comic1.append(name_of_comic[i])
    top_comic2.append(synopsis[i])
    top_comic3.append(genre[i])
    top_comic4.append(original_type[i])
    top_comic5.append(rating[i])
    
#แสดงออกมาประกอบไปด้วย อันดับ ค่าcosine similarity และชื่อเรื่องของcomic
```
