```python
import nltk
#import pickle

#nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
```

‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏≠‡∏≤‡πÑ‡∏ß‡πâ‡∏ó‡∏≥Query -> Word Token

```python
def get_token(doc_text):
  tokenizer = nltk.RegexpTokenizer("[\w']+")
  tokens = tokenizer.tokenize(doc_text)
  return tokens
```

‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏≠‡∏≤‡πÑ‡∏ß‡πâ‡∏ó‡∏≥Query -> Stop Word

```python
stopword=set(stopwords.words('english'))
def remove_stopword(doc_text):
  cleaned_text=[]
  for words in doc_text:
    if words not in stopword:
      cleaned_text.append(words)
  return cleaned_text   
```

‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏≠‡∏≤‡πÑ‡∏ß‡πâ‡∏ó‡∏≥Query -> Lemma

```python
def get_lemma(token_list):
  lem = nltk.stem.WordNetLemmatizer()
  lemma=[]
  for words in token_list:
    lemma.append(lem.lemmatize(words))
  return lemma
```

# Type Query ‚ùó
‡πÉ‡∏™‡πàquery ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡∏™‡∏∑‡∏ö‡∏Ñ‡πâ‡∏ô

```python
query = 'romantic king and queen'
```

‡∏ô‡∏≥query‡∏™‡πà‡∏á‡πÑ‡∏õ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥word token, stopword ‡πÅ‡∏•‡∏∞lemma

```python
query=get_token(query)
query=remove_stopword(query)
#print(query)
q=[]
for w in get_lemma(query):
  q.append(w)
q=' '.join(q)
```

‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤Cosine Similarity‡∏à‡∏≤‡∏ÅDoc ‡πÅ‡∏•‡∏∞ Query ‡πÇ‡∏î‡∏¢‡∏à‡∏∞print‡∏•‡∏≥‡∏î‡∏±‡∏ötop rank ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ 10 ‡∏≠‡∏±‡∏ô‡∏î‡∏±‡∏ö

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity = cosine_similarity(doc_1,query_1).flatten()
rank = cosine_similarity.argsort()[:-11:-1]
print(rank)
```
